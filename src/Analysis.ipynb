```python
import pandas as pd
import os
from utils import proportion_and_entropy_score_for_whole_dataset, restructure_data 
import geopandas as gpd
import altair as alt


```

```python
# read the data on country of birth and calculate the entropy 
data = pd.read_csv(os.path.join('..','data','country_of_birth_processed.csv')) 
data_cleaned = restructure_data(data, 'Analysis_region', 'Lower tier local authorities Code')
data_entropy = proportion_and_entropy_score_for_whole_dataset(data_cleaned)

# read the geo data in order to get the name of each local authority
data_geo = gpd.read_file(os.path.join('..','data','LAD2021.geojson')) 
data_geo = data_geo[['LAD21CD',	'LAD21NM']]

#merge the data
data_final = data_geo.merge(data_entropy, how = 'right', right_on = 'Lower tier local authorities Code', left_on = 'LAD21CD')
```

```python
# Examining correlation with Entropy
variable = ['Entropy',
            'Africa not otherwise specified and other',
            'Antarctica and Oceania',
            'Central America',
            'Central Asia',
            'Central and Western Africa',
            'Eastern Asia',
            'Middle East',
            'North Africa',
            'North America',
            'Other Europe',
            'South America',
            'South and Eastern Africa',
            'South-East Asia',
            'Southern Asia',
            'The Caribbean',
            'United Kingdom']
data_final[variable].corr().style.background_gradient(cmap='coolwarm')
```
